### 2.Write a python program to find the bill amount to be paid by the customer while ordering food online from a restaurant. The bill amount includes the amount for the food ordered based on the quantity and delivery charge based on distance of delivery as mentioned below:
### a. The Restaurant , home delivers the vegetarian combo costing Rs. 120 per plate and the non-vegetarian Rs. 150 per plate. Infact the restaurant gets more order for non-vegetarian combo than the vegetarian combo.
### b.The Customer must specify the cust_id, type of food, quantity (no. of paltes) required and the approximate distance in kms from the restaurant to the delivery point.
### The below information must be use to check the validity of data provided by the customer:

### · ‘V’ for vegetarian and ‘N’ for non-Vegetarian.

### · Distance in kms must be greater than 0.

### · Quantity ordered should be minimum 1

### Identify the cost of food and delivery charge based on the type of food and distance provided. Then the bill amount must be calculated as give below:

### Bill amount = cost per plate* quantity ordered+ Delivery charge

### Use the information provided in the table below to calculate the delivery charge. The bill amount should be returned as -1, if any of the inputs is invalid

#### Distance in kms.   --------                                              Delivery charge in Rs per km.

##### For first 3 kms -------                                                     0

##### For next 3 kms  --------                                                   3

##### For the remaining  -----------                                         6

#### Note: Create a CSV file for sample inputs for type of food, quantity and distance in kms with different values, and save the output in a JSON file.

```python
customer_id = input("Enter customer ID: ")
food_type = input("Enter food type (V for Vegetarian, N for Non-Vegetarian): ")
if food_type not in ['V', 'N']:
    print("Invalid food type entered!")
    exit()
quantity = int(input("Enter quantity required (number of plates): "))
if quantity < 1:
    print("Invalid quantity entered!")
    exit()
distance = float(input("Enter approximate distance from restaurant to delivery point (in km): "))
if distance <= 0:
    print("Invalid distance entered!")
    exit()

VEG_PRICE = 120
NONVEG_PRICE = 150
DELIVERY_CHARGE_1 = 0
DELIVERY_CHARGE_2 = 3
DELIVERY_CHARGE_3 = 6

if food_type == 'V':
    cost_per_plate = VEG_PRICE
else:
    cost_per_plate = NONVEG_PRICE
food_total = cost_per_plate * quantity

if distance <= 3:
    delivery_charge = DELIVERY_CHARGE_1
elif distance <= 6:
    delivery_charge = DELIVERY_CHARGE_2 * (distance - 3)
else:
    delivery_charge = DELIVERY_CHARGE_2 * 3 + DELIVERY_CHARGE_3 * (distance - 6)

bill_amount = food_total + delivery_charge

print("Customer ID:", customer_id)
print("Food type:", food_type)
print("Quantity required:", quantity)
print("Distance from restaurant to delivery point (in km):", distance)
print("Bill amount to be paid: Rs.", bill_amount)

import csv
import json


data = [
    ['101', 'V', 3, 7],
    ['105', 'N', 0, 5],
    ['102', 'N', 2, 15],
    ['107', 'V', 2, 2],
    ['111', 'N', 3, 6],
    ['121', 'N', 2, 3]
]

with open('input.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Customer ID', 'Food Type', 'Quantity', 'Distance'])
    for row in data:
        writer.writerow(row)


output = []
with open('input.csv', mode='r') as file:
    reader = csv.DictReader(file)
    for row in reader:
      
        quantity = int(row['Quantity'])
        distance = float(row['Distance'])

      
        if row['Food Type'] == 'V':
            cost_per_plate = 120
        else:
            cost_per_plate = 150
        food_total = cost_per_plate * quantity

       
        if distance <= 3:
            delivery_charge = 0
        elif distance <= 6:
            delivery_charge = 3 * (distance - 3)
        else:
            delivery_charge = 3 * 3 + 6 * (distance - 6)

  
        bill_amount = food_total + delivery_charge

        
        output.append({
            'Customer ID': row['Customer ID'],
            'Food Type': row['Food Type'],
            'Quantity': quantity,
            'Distance': distance,
            'Bill Amount': bill_amount
        })


with open('output.json', mode='a') as file:
    json.dump(output, file, indent=2)

print('Output saved to output.json')
```
