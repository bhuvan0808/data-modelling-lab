### 2.From Question 1. Read numbers.txt If it is a number, check whether the number is a Strong numbers or not. If its a strong number, save it in another file called strong_num.txt.

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
def is_strong_number(n):
    sum_of_factorials = sum(factorial(int(digit)) for digit in str(n))
    return sum_of_factorials == n
with open('numbers.txt', 'r') as f, open('strong_num.txt', 'w') as output:
    for line in f:
        num = line.strip()
        if num.isdigit() and is_strong_number(int(num)):
            output.write(num + '\n')
```

### 3.A file (question_txt) contains Integer values where each line represents a list, write a python program to check whether it contains same number in subsequent position. Display the count of such occurrences and store the result in a csv file (as shown below).

```python
import csv

input_file = 'question_3 .txt'
output_file = 'output.csv'

with open(input_file, 'r') as f:
    lines = f.readlines()

with open(output_file, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Line', 'Count'])

    for i, line in enumerate(lines, start=1):
        line = line.strip()

        if line:  # Skip empty lines
            nums = [int(n) for n in line.split(',')]
            count = 0

            for j in range(len(nums)-1):
                if nums[j] == nums[j+1]:
                    count += 1

            writer.writerow([i, count])

print("Results saved to", output_file)
```
